```python
import numpy as np

print( 'The Mean for Last_evaulation is: ', 
      HRM['last_evaluation'].mean())

print( 'The Median for Last_evaulation is: ', 
      HRM['last_evaluation'].median())

print( 'The Mode for Last_evaulation is: ', 
      HRM['last_evaluation'].mode())

print( 'The Minimum for Last_evaulation is: ', 
      HRM['last_evaluation'].min())

print( 'The Maximum for Last_evaulation is: ', 
      HRM['last_evaluation'].max())

print( 'The Variance for Last_evaulation is: ', np.var(HRM['last_evaluation']))

```
