```python
## Loading of the Data 

import pandas as pd

air_passengers = pd. read_csv('AirPassengers.csv', header=0, parse_dates=[0], index_col=0)
```

```python
## Visualizing the Time Series

from matplotlib import pyplot
pyplot.figure(figsize=(10,5), dpi=100 )
pyplot.plot(air_passengers, color='tab:orange')  
pyplot.gca().set(title="Number of air passengers over time", 
            xlabel='Month', ylabel='Passengers')
pyplot.show()
```
